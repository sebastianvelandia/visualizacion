```python
df1=df1.dropna(how='any')
```

```python
df['sex']=df['sex'].replace({'female':0,'male':1});

farePerSex = df.groupby(['sex'],as_index=False)['fare'].sum();
sex = farePerSex['sex'];
fare = farePerSex['fare'];

#sex['sex']=df['sex'].replace({'0':'female','1':'male'});

print(farePerSex)

plt.barh(sex, fare);
```

```python
#seperating independent and dependent variables
y = df['survived']
X = df.drop(['survived'], axis=1)
```

```python
#importing train_test_split to create validation set
from sklearn.model_selection import train_test_split
```

```python
#creating the train and validation set
X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state = 101, stratify=y, test_size=0.25)
```

```python
df.info()
```

```python
from sklearn import tree
```

```python
decision_tree = tree.DecisionTreeClassifier(criterion='gini',max_depth=4,class_weight="balanced")
decision_tree.fit(X_train, y_train)

# exportar el modelo a archivo .dot
with open(r"tree1.dot", 'w') as f:
     f = tree.export_graphviz(decision_tree,
                              out_file=f,
                              max_depth = 7,
                              impurity = True,
                              feature_names = list(df.drop(['survived'], axis=1)),
                              class_names = ['0', '1'],
                              rounded = True,
                              filled= True )
        
# Convertir el archivo .dot a png para poder visualizarlo

check_call(['dot','-Tpng',r'tree1.dot','-o',r'tree1.png'])
PImage("tree1.png")
```

```python
decision_tree.score(X_train, y_train)
```

```python
decision_tree.score(X_valid, y_valid)
```

```python

from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,10))
plot_tree(decision_tree=decision_tree,feature_names=X_train.columns,filled=True,fontsize=8,max_depth=4,);
plt.savefig('Arbol.png')
```
